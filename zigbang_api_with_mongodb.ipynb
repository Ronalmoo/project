```python
query = {}
projection = {
    "_id": False,
    "address1": True,
    "address2": True,
    "agent_lat": True,
    "agent_lng": True,
    "agent_name": True,
    "agent_phone": True,
    "deposit": True,
    "elevator": True,
    "rent": True,
    "floor": True,
    "manage_cost": True,
    "options": True,
    "view_count": True
}
result = list(zigbang.find(filter=query, projection=projection))
```

```python
items[0].keys()
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(result)
df
```

```python
sorted_df = df.sort_values("view_count", ascending=False)
sorted_df
```

```python
sorted_df['manage_cost'] = sorted_df['manage_cost'].apply(
    lambda cost : 0 if cost == "없음" else cost.replace("만원", "")
)
sorted_df.tail()
```

```python
 from ipyleaflet import *
```

```python
def draw_map(addr="잠원동"):
    center = zb.get_addr_info(addr)
    print(center)
    return Map(center=center, zoom=15)
```

```python
map_obj = draw_map("잠원동")
map_obj
```

```python
locations = list(zip(sorted_df['agent_lat'], sorted_df['agent_lng']))
locations[:2]
```

```python
import time
for location in locations:
    marker = Marker(location=location)
    map_obj += marker
    time.sleep(0.2)
```
